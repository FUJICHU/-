```python
import pandas as pd 
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号'-'显示为方块的问题
```

```python
df = pd.read_csv('clean.csv')
# 使用 groupby 聚合函数计算每个区域的房屋数量和平均价格
df_stats = df.groupby('区域').agg({'总价（万）': 'count', '单价': 'mean'})

 # 重置索引并更改列名
df_stats = df_stats.reset_index().rename(columns={'区域': '城市名', '总价（万）': '房屋数量', '单价': '平均价格'})
# 获取城市名和平均价格
city_names = df_stats['城市名']
average_prices = df_stats['平均价格']

# 绘制柱状图
plt.figure(figsize=(10, 6))
plt.bar(city_names, average_prices, color='skyblue')
plt.xlabel('城市名')
plt.ylabel('平均价格')
plt.title('各区域二手房平均价格')
plt.xticks(rotation=45)  # 使x轴标签倾斜，以防止重叠
plt.show()
```

```python
df = pd.read_csv('clean.csv')

# 去除建造时间列为空的行
data = df.dropna(subset=['建造时间'])

# 提取年份和总价列
data.loc[:, '年份'] = data['建造时间'].astype(str).str[:4].astype(int)
data.loc[:, '总价（万）'] = data['总价（万）'].astype(float)

# 计算每年的平均价格
average_prices = data.groupby('年份')['总价（万）'].mean().reset_index()

# 获取年份和平均价格
years = average_prices['年份']
average_prices_values = average_prices['总价（万）']

# 绘制柱状图
plt.figure(figsize=(12, 6))
plt.bar(years, average_prices_values, color='green')
plt.xlabel('年份')
plt.ylabel('平均价格')
plt.title('不同年份二手房平均价格对比')
plt.xticks(rotation=90)  # 使x轴标签倾斜，以防止重叠
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('clean.csv')
v_list = list(df['房间数'].value_counts().sort_values(ascending=False))[:30]
k_list = list(df['房间数'].value_counts().sort_values(ascending=False).index)[:30]
result = []
for item in range(len(v_list)):
    result.append((v_list[item], k_list[item]))

xiangxing = []
for n in k_list:
    xiangxing.append(list(df[df['房间数'] == n]['单价']))

# 绘制散点图
plt.figure(figsize=(8, 6))
for i in range(len(k_list)):
    plt.scatter([i] * len(xiangxing[i]), xiangxing[i], label=k_list[i])
plt.xticks(range(len(k_list)), k_list)
plt.xlabel('房间数')
plt.ylabel('单价')
plt.title('不同房间数与价格的散点图')
plt.legend()
plt.show()
```

```python
df = pd.read_csv('clean.csv')
v_list = list(df['朝向'].str.split(' ', expand=True)[0].value_counts())
k_list = list(df['朝向'].str.split(' ', expand=True)[0].value_counts().index)
result = []
for item in range(len(v_list)):
    result.append((v_list[item], k_list[item]))
xiangxing = []
for n in k_list:
    xiangxing.append(list(df[df['朝向'] == n]['单价']))
# 绘制箱形图
plt.figure(figsize=(8, 6))
plt.boxplot(xiangxing, labels=k_list)
plt.xlabel('朝向')
plt.ylabel('单价')
plt.title('不同朝向的单价分布情况')
plt.show()
```

```python
df = pd.read_csv('clean.csv')


# 根据房子大小分组并计算每个区间的平均价格
bins = [0, 80, 100, 120, 140, 160, float('inf')]  # 自定义区间
labels = ['0-80', '80-100', '100-120', '120-140', '140-160', '160+']  # 区间标签
df['房子大小区间'] = pd.cut(df['大小'], bins=bins, labels=labels, right=False)  # 将房子大小分为不同区间

result = df.groupby('房子大小区间')['总价（万）'].mean()  # 计算每个区间大小的房子的平均价格

# 绘制柱状图
plt.bar(result.index, result.values)
plt.xlabel('房子大小区间')
plt.ylabel('平均价格（万）')
plt.title('房子大小与价格关系的柱状图')
plt.show()
```
